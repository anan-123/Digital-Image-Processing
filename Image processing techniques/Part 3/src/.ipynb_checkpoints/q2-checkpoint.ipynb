```python
def erosion(image, size):
    img = image.copy()
    kernel = np.ones((size,size), dtype=np.uint8)
    img_op = np.zeros(img.shape)
    n = size // 2
      
    for r in range(n,img.shape[0]-n):
      for c in range(n,img.shape[1]-n):
            img_op[r][c] = np.min(img[r-n:r+n+1,c-n:c+n+1]*kernel)
  
    return img_op.astype(np.uint8)
```

```python
def dilation(image, size):
    img = image.copy()
    kernel = np.ones((size,size), dtype=np.uint8)
    img_op = np.zeros(img.shape)
    n = size // 2
      
    for r in range(n,img.shape[0]-n):
      for c in range(n,img.shape[1]-n):
            img_op[r][c] = np.max(img[r-n:r+n+1,c-n:c+n+1]*kernel)
  
    return img_op.astype(np.uint8)
```

```python
def threshold(img,k):
    im = img.copy()
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            if im[i][j]>k:
                im[i][j]=255
            else:
                im[i][j]=0
    return im
```

```python
img = cv2.imread('../images/cell.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = threshold(img,150)
img1 = erosion(img,1)
plt.imshow(img1,cmap='gray')
```

```python
def BFS(img,i,j,c):
        queue = []
        queue.append((i,j))
        visited[(i,j)] = 1
        colors[i][j]=c
        while queue:
            a,b = queue.pop(0)
            for p in range(-1,2):
                for q in range(-1,2):
                    if (a+p) >=0 and (a+p) <img.shape[0] and (b+q) >=0 and (b+q) <img.shape[1]:
                        if img[a+p][b+q]==255 and visited[a+p][b+q] == 0:
                            queue.append((a+p,b+q))
                            colors[a+p][b+q] = c
                            visited[(a+p,b+q)] = 1
```

# PART 2

```python
image = cv2.imread("../images/objects.png",0)
im = threshold(image,127)
plt.imshow(im,cmap='gray')
```

```python
def DFS(i,j,image1,result,visited):
    if i<0 or i>=image1.shape[0] or j<0 or j>=image1.shape[1] or visited[i,j] :
        return 0
    visited[i][j] = 1
    if (image1[i][j]):
        result[i,j] = 1
        for p in range(-1,2):
            for q in range(-1,2):
                if p !=q: 
                    DFS(i+p,j+q,image1,result,visited)
    return 0
```

```python
def component_count(img1,img2):    
    c = 0
    visited = np.zeros(img1.shape)
    result = np.zeros(img1.shape,dtype=np.uint16)
    for i in range(img1.shape[0]):
        for j in range(img1.shape[1]):
            if img1[i,j] and visited[i,j] == 0:
                    DFS(i,j,img2,result,visited)
                    c = c+1
    return result,c
```

```python
def fillhole(im,SE):
    k = SE.shape[0]//2
    result = np.zeros(im.shape) 
    padded_img = np.pad(im,((k,k),(k,k)),'constant')
    for i in range(k,im.shape[0]+k):
        for j in range(k,im.shape[1]+k):
            arr = np.equal(padded_img[i-k:i+k+1,j-k:j+k+1],SE)
            if np.sum(arr) == SE.size:
                result[i-k][j-k] = 1
    return result.astype(np.uint8)
```

```python
img1 = cv2.imread("../images/objects.png",0)
img1 = (img1>127).astype(np.uint8)
```

```python
img2 = cv2.dilate(img1,np.ones((7,9)),iterations=1)
img2 = cv2.erode(img2,np.ones((7,9)),iterations=1)
plt.imshow(img2,cmap='gray')
hole_img = img2 - img1
hole_img = cv2.erode(hole_img,np.ones((6,6)),iterations=1)
hole_img = cv2.dilate(hole_img,np.ones((6,6)),iterations=1)
```

# How many objects have one or more holes?

```python
holes,count = component_count(hole_img ,img2)
result_holes = np.logical_and(holes,img1)
print("Number of objects with holes",count)
plt.imshow(result_holes.astype(np.uint8),cmap='gray')
```

# How many square objects are in the image?

```python
sq = fillhole(img2,np.array([[0,0,0],[0,1,1],[0,1,1]]))
result_square,count = component_count(sq,img2)
print("No of squares",count)
plt.imshow(result_square,cmap='gray')
```

# Identify the square objects that have holes.

```python
hole_square = np.logical_and(result_square,result_holes)
square,count = component_count(hole_square,hole_square)
print("No of square with holes",count)
plt.imshow(square,cmap='gray')
```

#  Identify the circular objects that have no holes

```python
without_holes = img1 - result_holes
circle = fillhole(without_holes ,np.array([[1,1,0],[1,1,0],[1,0,0]]))
final_circle,count = component_count(circle,without_holes)
print("No of circles with no holes = ",count)
plt.imshow(final_circle,cmap='gray')
```

```python
def BFS2(img,i,j,c):
        queue = []
        queue.append((i,j))
        visited[(i,j)] = 1
        colors[i][j]=c
        while queue:
            a,b = queue.pop(0)
            for p in range(-1,2):
                for q in range(-1,2):
                    if (a+p) >=0 and (a+p) <img.shape[0] and (b+q) >=0 and (b+q) <img.shape[1]:
                        if img[a+p][b+q]==255 and visited[a+p][b+q] == 0:
                            if (a+p)==0 or (a+p)==(img.shape[0]-1) or (b+q)==0 or (b+q)== (img.shape[1]-1):
                                boundary_colors.append(c)
                            queue.append((a+p,b+q))
                            colors[a+p][b+q] = c
                            visited[(a+p,b+q)] = 1
```

```python
visited = np.zeros(img.shape)
colors = np.zeros(img.shape)
boundary_colors = []
c = 0
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i][j]==255 and visited[i][j] == 0:
            if i==0 or i==(img.shape[0]-1) or j==0 or j== (img.shape[1]-1):
                boundary_colors.append(c)
            BFS2(img,i,j,c)
            c = c+1
            
print(c)
```

# Coins overlapping

```python
img = cv2.imread('../images/circles.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img = threshold(img,150)
#plt.imshow(img,cmap = 'gray')
```

```python
def BFS3(img,i,j,c):
        area = 1
        queue = []
        queue.append((i,j))
        visited[(i,j)] = 1
        colors[i][j]=c
        while queue:
            a,b = queue.pop(0)
            for p in range(-1,2):
                for q in range(-1,2):
                    if (a+p) >=0 and (a+p) <img.shape[0] and (b+q) >=0 and (b+q) <img.shape[1]:
                        if img[a+p][b+q]==255 and visited[a+p][b+q] == 0:
                            queue.append((a+p,b+q))
                            area = area +1 
                            colors[a+p][b+q] = c
                            visited[(a+p,b+q)] = 1
        return area
```

```python
visited = np.zeros(img.shape)
colors = np.zeros(img.shape)
overlap_colors = []
c = 0
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i][j]==255 and visited[i][j] == 0:
            l = BFS3(img,i,j,c)
            if l > 450:
                overlap_colors.append(c)
            c = c+1
print(c)
```

```python
output = np.zeros((img.shape[0],img.shape[1]))
colors = colors.astype(np.uint8)
COLORS = [(0,0,0),(255, 0, 0), (0, 255, 0), (0,0, 255), (10, 78, 0), (10,120,14), (86,85,11), (25,10, 25), (24,85,23), (96,98,16)]
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
                c = colors[i][j]
                if c in overlap_colors and img[i][j]==255:
                    output[i][j] = 255
                else:
                    output[i][j]=0
                    
        
plt.imshow(output,cmap='gray')
```

# Non overlapping coins

```python
output = np.zeros((img.shape[0],img.shape[1]))
colors = colors.astype(np.uint8)
COLORS = [(0,0,0),(255, 0, 0), (0, 255, 0), (0,0, 255), (10, 78, 0), (10,120,14), (86,85,11), (25,10, 25), (24,85,23), (96,98,16)]
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
                c = colors[i][j]
                if c not in overlap_colors and img[i][j]==255:
                    output[i][j] = 255
                else:
                    output[i][j]=0
                    
        
plt.imshow(output,cmap='gray')
```
