# Part 1: thresholding

```python
def threshold(img,k):
    im = img.copy()
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            if im[i][j]>k:
                im[i][j]=255
            else:
                im[i][j]=0
    return im
```

```python
listk = [100,127,130,140,150,180]
plt.figure(figsize=(20,20))
i = 1;
for k in listk:
    output_img = threshold(img,k)
    plt.subplot(3,3,i)
    i = i+1
    plt.imshow(output_img,cmap='gray')
    plt.title('Threshold = '+str(k))
```

```python
def erosion(image, size):
    img = image.copy()
    kernel = np.ones((size,size), dtype=np.uint8)
    img_op = np.zeros(img.shape)
    n = size // 2
      
    for r in range(n,img.shape[0]-n):
      for c in range(n,img.shape[1]-n):
            img_op[r][c] = np.min(img[r-n:r+n+1,c-n:c+n+1]*kernel)
  
    return img_op.astype(np.uint8)
```

```python
def dilation(image, size):
    img = image.copy()
    kernel = np.ones((size,size), dtype=np.uint8)
    img_op = np.zeros(img.shape)
    n = size // 2
      
    for r in range(n,img.shape[0]-n):
      for c in range(n,img.shape[1]-n):
            img_op[r][c] = np.max(img[r-n:r+n+1,c-n:c+n+1]*kernel)
  
    return img_op.astype(np.uint8)
```

```python
img = cv2.imread('../images/coins.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python

img1 = threshold(img,127) #ideal threshold for morphological operations
img2 = dilation(img1,9)
img3 = erosion(img2,21)
plt.imshow(img3,cmap='gray')
```

# no of coins

```python
visited = np.zeros(img.shape)
```

```python
def BFS(img,i,j):
        queue = []
        queue.append((i,j))
        visited[(i,j)] = 1

        while queue:
            a,b = queue.pop(0)
            for p in range(-1,2):
                for q in range(-1,2):
                    if img[a+p][b+q]==255 and visited[a+p][b+q] == 0:
                        queue.append((a+p,b+q))
                        visited[(a+p,b+q)] = 1
            
            
```

```python
c = 0
for i in range(img3.shape[0]):
    for j in range(img3.shape[1]):
        if img3[i][j]==255 and visited[i][j] == 0:
            BFS(img3,i,j)
            c = c+1
print(c)
            
```

# OTSU thresholding


```python
hist,edges = np.histogram(img,bins= 256)
hist = np.divide(hist.ravel(),hist.max())
w1 = np.cumsum(hist)
w2 = np.cumsum(hist[::-1])[::-1]
mids = (edges[:-1] + edges[1:]) / 2.
mean1 = np.cumsum(hist * mids) / w1
mean2 = (np.cumsum((hist * mids)[::-1]) / w2[::-1])[::-1]
index = np.argmax(w1[:-1] * w2[1:] * (mean1[:-1] - mean2[1:]) ** 2)
t = mids[:-1][index]
print(t)
```

```python
img = cv2.imread('../images/coins.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

img1 = np.where(img>t,255,0)
plt.imshow(img1,cmap='gray')
```

```python
visited = np.zeros(img.shape)
c = 0
for i in range(img3.shape[0]):
    for j in range(img3.shape[1]):
        if img3[i][j]==255 and visited[i][j] == 0:
            BFS(img3,i,j)
            c = c+1
print(c)
            
```

The number of coins in a and b are different, it is 24 in first case and 21 in second. In otsu's thresholding we don't need to perform morphological operations to get close to our correct answer. The noise affects the answer but its less as compared to when we apply thresholding (before doing morphological operations)

```python
img = cv2.imread('../images/coins.png')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

```python
def convolution(image,kernel):
    img=image.copy()
    img=img.astype(np.float64)
    kernel= kernel.astype(np.float64)
    output_img = np.zeros(img.shape)
    for i in range(int(kernel.shape[0]/2),image.shape[0]-int((kernel.shape[0]-1)/2)):
        for j in range(int(kernel.shape[1]/2),image.shape[1]-int((kernel.shape[1]-1)/2)):
            output_img[i][j]=sum(sum(kernel* img[i-int(kernel.shape[0]/2) : i+1+int((kernel.shape[0]-1)/2),j-int(kernel.shape[1]/2):j+1+int((kernel.shape[1]-1)/2)]))
            #checking bounds
            output_img[i][j] = min(255,output_img[i][j])
            output_img[i][j] = max(0,output_img[i][j])
    output_img = output_img.astype(np.uint8)
    return output_img
```

```python
def sobel(image):
    img=image.copy()
    kernel = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]) 
    output_img1 = convolution(img,kernel)
    kernel = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
    output_img2 = convolution(img,kernel)
  
    output_img1 = output_img1.astype(np.float64)
    output_img2 = output_img2.astype(np.float64)
    output_img = np.sqrt(output_img1*output_img1+output_img2*output_img2)
    output_img=output_img.astype(np.uint8)

    plt.imshow(output_img,cmap='gray')
    plt.title('sobel')
    return output_img
```
