```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Reading 4 images

```python
img1 = cv2.imread('../images/bike.jpg',cv2.IMREAD_UNCHANGED)
img2 = cv2.imread('../images/chess1.jpg',cv2.IMREAD_UNCHANGED)
img3 = cv2.imread('../images/logo.jpg',cv2.IMREAD_UNCHANGED)
img4 = cv2.imread('../images/waffle.jpg',cv2.IMREAD_UNCHANGED)
fig = plt.figure(figsize=(15, 8))
fig.add_subplot(2,2,1)
plt.imshow(img1)
fig.add_subplot(2,2,2)
plt.imshow(img2)
fig.add_subplot(2,2,3)
plt.imshow(img3)
fig.add_subplot(2,2,4)
plt.imshow(img4)
```

# The above images are in BGR format since opencv reads images in bgr format. Converting to rgb format using cvtColor resolves the issue.

```python
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(15, 8))
fig.add_subplot(2,2,1)
plt.imshow(img1)
fig.add_subplot(2,2,2)
plt.imshow(img2)
fig.add_subplot(2,2,3)
plt.imshow(img3)
fig.add_subplot(2,2,4)
plt.imshow(img4)
```

# Printing the dimensions of images

```python
print(img1.shape)
height,width,channels = img1.shape

print(height)
print(width)
print(channels)
```

# Converting image to greyscale
Since there are three channels the average of them will give grey scale.

```python

for i in range(height):
    for j in range(width):
        img1[i, j] = sum(img1[i, j]) * 0.33
cv2.imwrite('../images/bike_greyscale.jpg', img1)
```

# Range of values of image

```python
print(np.max(img3[:,:,0]))
print(np.max(img3[:,:,1]))
print(np.max(img3[:,:,2]))
```

The range of pixel values is from 0(no color) to 255(full color) (8 bit) and the color image has 3 channels (rgb) .
