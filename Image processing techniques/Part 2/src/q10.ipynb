```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rotImage(image, angle):
  img = image.copy()
  output_img = np.zeros_like(img)
  r = np.array([[np.cos(np.radians(angle)), np.sin(np.radians(angle)), 0],[-np.sin(np.radians(angle)), np.cos(np.radians(angle)), 0],[0, 0, 1]])
  t = np.array([[1, 0,img.shape[1]// 2],[0, 1,img.shape[0]//2],[0, 0, 1]])
  #scaling by 0.75 so that the image fits
  s = np.array([[0.75, 0, 0],[0,0.75, 0],[0, 0, 1]])
  mat = t@r@s@np.linalg.inv(t)
  mat_inv = np.linalg.inv(mat)

  # Grid
  grid = np.indices((img.shape[1], img.shape[0])).reshape(2, -1)
  grid = np.vstack((grid, np.ones(grid.shape[1]))).astype(np.int)

  # transform, nearest neighbour interpolate
  warpgrid = np.round(mat@grid)
  warpgrid = warpgrid.astype(np.int)

  p = grid[0]
  q = grid[1]
  r = warpgrid[0, :]
  s = warpgrid[1, :]
  #get the pixels that are inside the gird
  index = np.where((r >= 0) & (r < img.shape[1]) &(s >= 0) & (s< img.shape[0]))
  output_img[s[index],r[index]] = img[q[index],p[index]]
  return output_img
```

```python
angle = [15,30,45,60,75,90]
```

```python
img = cv2.imread('../images/eye.png',cv2.IMREAD_UNCHANGED)
img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(50,25))
plt.subplot(len(angle)+1,1,1)
i = 2
plt.imshow(img)
plt.title('original image')
for a in angle:
    output_img1 = rotImage(img,a)
    plt.subplot(len(angle)+1,1,i)
    i = i+1
    plt.imshow(output_img1)
    plt.title('angle = '+str(a))
```
