```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../images/mountain.png',cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img,(512,512))
plt.imshow(img,'gray')
```

```python
def FFT_1D(f):
    f1 = f.copy()
    f1 = np.asarray(f1,dtype=complex)
    if f1.shape[0]%2!= 0:
        return 
    if f1.shape[0]<=4:
        return DFT_1D(f1)
    else:
        f1_even = FFT_1D(f1[::2])  # even part
        f1_odd = FFT_1D(f1[1::2])  # odd part
        n = f1.shape[0]
        W = np.exp(-2j*np.pi*np.arange(n)/n)
        k = n//2
        opf = np.concatenate([f1_even + f1_odd * W[:k],f1_even + f1_odd * W[k:]])

    return opf
```

```python
def DFT_1D(f):
    f1 = f.copy()
    f2 = np.asarray(f,dtype=complex)
    sum =0 
    for i in range(f1.shape[0]):
        for j in range(f1.shape[0]):
            sum = sum + f2[j]*np.exp(-2j*np.pi*j*i*np.divide(1,f1.shape[0],dtype=complex))
        f1[i] = sum
        sum=0
    return f1
```

```python
def FFT_2D(f):
    f1 = f.copy()
    opf = np.zeros(f1.shape, dtype=complex)
    if len(f1.shape)==3: #colored image
        for k in range(3):
            opf[:,:,k]=FFT_2D(f1[:,:,k])
    if len(f1.shape) == 2: #grayscale or one channel
        for i in range(f1.shape[0]):opf[i,:]=FFT_1D(f1[i,:])
        for j in range(f1.shape[1]):opf[:,j]=FFT_1D(opf[:,j])
            
    return opf
```

# Low pass filter

```python
def lowpassfilter(fft_img,x):
  temp= np.zeros(fft_img.shape)
  for r in range(fft_img.shape[0]):
    for c in range(fft_img.shape[1]):
      if (abs(r -fft_img.shape[0]//2)**2 + abs(c- fft_img.shape[1]//2)**2) <= x**2:
        temp[r][c] = 1
  return temp * fft_img
```

```python
img = cv2.imread("../images/lena.jpg",cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img,(256,256))
img= np.pad(img,((0,img.shape[0]),(0,img.shape[1])),mode='constant')
img= img.astype('int16')
for r in range(img.shape[0]):
    for c in range(img.shape[1]):
      if (r+c)%2:
        img[r][c] = -1 * img[r][c]
fft_img = FFT_2D(img)
```

```python
fft_lowpass = lowpassfilter(fft_img,50)
opt = np.log(1 + np.abs(fft_lowpass))
plt.imshow(opt,cmap='gray')
plt.title('D = 50')
```

```python
fft_lowpass = lowpassfilter(fft_img,15)
opt = np.log(1 + np.abs(fft_lowpass))
plt.imshow(opt,cmap='gray')
plt.title('D = 15')
```

```python
fft_lowpass = lowpassfilter(fft_img,30)
opt = np.log(1 + np.abs(fft_lowpass))
plt.imshow(opt,cmap='gray')
plt.title('D = 30')
```

# Gaussian low pass filter

```python
def gaussianlowpass(fft_img,x):
  temp= np.zeros(fft_img.shape)
  for r in range(fft_img.shape[0]):
    for c in range(fft_img.shape[1]):
        temp[r][c]=np.exp(-1*((abs(r -fft_img.shape[0]//2)**2 + abs(c- fft_img.shape[1]//2)**2)/(2*(x**2))))
        
  return fft_img*temp
 
```

```python
img = cv2.imread("../images/lena.jpg",cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img,(256,256))
img= np.pad(img,((0,img.shape[0]),(0,img.shape[1])),mode='constant')
img= img.astype('int16')
for r in range(img.shape[0]):
    for c in range(img.shape[1]):
      if (r+c)%2:
        img[r][c] = -1 * img[r][c]
fft_img = FFT_2D(img)
```

```python
fft_lowpass = gaussianlowpass(fft_img,50)
opt = np.log(1 + np.abs(fft_lowpass))
plt.imshow(opt,cmap='gray')
plt.title('D = 50')
```

```python
fft_lowpass = gaussianlowpass(fft_img,15)
opt = np.log(1 + np.abs(fft_lowpass))
plt.imshow(opt,cmap='gray')
plt.title('D = 15')
```

```python
fft_lowpass = gaussianlowpass(fft_img,30)
opt = np.log(1 + np.abs(fft_lowpass))
plt.imshow(opt,cmap='gray')
plt.title('D = 30')
```

# Observations

Ideal low pass filter keeps all frequencies below a limit and omit others. So if we increase D the number of frequencies included are more.  Gaussian low pass filter doesn't cut frequency sharply like ideal low pass filter. It has a smooth frequency boundary. As the D size increases more frequencies are included.
