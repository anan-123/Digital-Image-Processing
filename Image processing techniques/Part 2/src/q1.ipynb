# Low pass filtering - average filter

```python
def convolution(image,kernel):
    img=image.copy()
    img=img.astype(np.float64)
    kernel= kernel.astype(np.float64)
    output_img = np.zeros(img.shape)
    for i in range(int(kernel.shape[0]/2),image.shape[0]-int((kernel.shape[0]-1)/2)):
        for j in range(int(kernel.shape[1]/2),image.shape[1]-int((kernel.shape[1]-1)/2)):
            output_img[i][j]=sum(sum(kernel* img[i-int(kernel.shape[0]/2) : i+1+int((kernel.shape[0]-1)/2),j-int(kernel.shape[1]/2):j+1+int((kernel.shape[1]-1)/2)]))
            #checking bounds
            output_img[i][j] = min(255,output_img[i][j])
            output_img[i][j] = max(0,output_img[i][j])
    output_img = output_img.astype(np.uint8)
    return output_img
```

```python
img = cv2.imread('../images/ceramic.jpg',cv2.IMREAD_GRAYSCALE)
listk = [3,5,7,11,13]
plt.figure(figsize=(50,40))
i = 1;
for k in listk:
    kernel = np.ones((k,k))/k**2
    output_img = convolution(img,kernel)
    plt.subplot(len(listk),1,i)
    i = i+1
    plt.imshow(output_img,cmap='gray')
    plt.title('k = '+str(k))
```

# Efficient average filter

```python
def efficient_average_filter(image,k):
    img=image.copy()
    img=img.astype(np.float64)
    cumy = np.zeros(img.shape)
    cumx=np.zeros(img.shape)
    for i in range(img.shape[0]):
            cumx[i][0]=img[i][0]
            for j in range(1,img.shape[1]):
                cumx[i][j]=cumx[i][j-1]+img[i][j]
                if j>=k:
                    cumx[i][j]=cumx[i][j]-img[i][j-k]
    for j in range(img.shape[1]):
            cumy[0][j]=cumx[0][j]
            for i in range(1,img.shape[0]):
                cumy[i][j]=cumy[i-1][j]+cumx[i][j]
                if i>=k:
                    cumy[i][j]=cumy[i][j]-cumx[i-k][j]
        
    output_img = np.zeros(img.shape)
    n = k//2
    for i in range(n,img.shape[0]-n):
        for j in range(n,img.shape[1]-n):
            output_img[i][j] = cumy[i+n][j+n]/(k*k)
    output_img = output_img.astype(np.uint8)
    return output_img
```

```python
img = cv2.imread('../images/ceramic.jpg',cv2.IMREAD_GRAYSCALE)
listk = [3,5,7,11,13]
plt.figure(figsize=(50,40))
i = 1;
for k in listk:
    kernel = np.ones((k,k))/k**2
    output_img = efficient_average_filter(img,k)
    plt.subplot(len(listk),1,i)
    i = i+1
    plt.imshow(output_img,cmap='gray')
    plt.title('k = '+str(k))
```

# Median filter

```python
def median_filter(image,k):
    n=k//2
    img=image.copy()
    img=img.astype(np.float64)
    output_img = np.zeros(img.shape)
    for i in range(n,img.shape[0]-n):
        for j in range(n,img.shape[1]-n):
            list_pixels=[]
            for p in range(i-n,i+n+1):
                for q in range(j-n,j+n+1):
                    list_pixels.append(image[p][q])
            list_pixels.sort()
            median_element = int(len(list_pixels) / 2)
            output_img[i][j] = list_pixels[median_element]
    output_img = output_img.astype(np.uint8)
    return output_img
```

```python
img = cv2.imread('../images/ceramic.jpg',cv2.IMREAD_GRAYSCALE)
listk = [3,5,7,11,13]
plt.figure(figsize=(50,40))
i = 1;
for k in listk:
    kernel = np.ones((k,k))/k**2
    output_img = median_filter(img,k)
    plt.subplot(len(listk),1,i)
    i = i+1
    plt.imshow(output_img,cmap='gray')
    plt.title('k = '+str(k))
```
