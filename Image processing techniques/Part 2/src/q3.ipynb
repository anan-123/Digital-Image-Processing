```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolution(image,kernel):
    img=image.copy()
    img=img.astype(np.float64)
    kernel= kernel.astype(np.float64)
    output_img = np.zeros(img.shape)
    for i in range(int(kernel.shape[0]/2),image.shape[0]-int((kernel.shape[0]-1)/2)):
        for j in range(int(kernel.shape[1]/2),image.shape[1]-int((kernel.shape[1]-1)/2)):
            output_img[i][j]=sum(sum(kernel* img[i-int(kernel.shape[0]/2) : i+1+int((kernel.shape[0]-1)/2),j-int(kernel.shape[1]/2):j+1+int((kernel.shape[1]-1)/2)]))
            #checking bounds
            output_img[i][j] = min(255,output_img[i][j])
            output_img[i][j] = max(0,output_img[i][j])
    output_img = output_img.astype(np.uint8)
    return output_img
```

# Gaussian filter

```python
def gaussfilter(image,sigma,size):
    img=image.copy()
    n = size // 2
    values = np.zeros((size, size), dtype=tuple)
    for x in range(0, 2*n+1):
      for y in range(0, 2*n+1):
        values[x,y] = (x-n,y-n)
    kernel = np.zeros((size, size))
    for r in range(size):
        for c in range(size):
              x,y = values[r][c] 
              kernel[r][c] = (1/(2*np.pi*(sigma**2)))*np.exp(-(x**2 + y**2)/(2* sigma**2))
    outputimg=convolution(img,kernel)
    return outputimg
                                                 
```

```python
size_sigma = [(5,2/3),(5,1),(5,2),(7,1),(9,4/3)]
```

```python
img = cv2.imread('../images/mountain.png',cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(100,50))
plt.subplot(len(size_sigma)+1,1,1)
i = 2
plt.imshow(img,cmap='gray')
plt.title('original image')
for v in size_sigma:
    sigma = v[1]
    size = v[0]
    output_img1 = gaussfilter(img,sigma,size)
    plt.subplot(len(size_sigma)+1,1,i)
    i = i+1
    plt.imshow(output_img1,cmap='gray')
    plt.title('sigma = '+str(sigma)+'size = '+str(size))
```

# Observations

Since 99 percent of gaussian = 6 `*` sigma so the ideal sigma for kernel size n is around (n-1)/6.
so for kernel size: 5 ideal sigma = 2/3 = 0.6666
similarly for other kernel size. We can see that when sigma is more than this number the image becomes blurred and when its less its sharp.

# Bilateral filter

```python
def bilateral_filter(image, sigmad, sigmar, size):
    sz=size
    n = size//2
    img=image.copy()
    img = img.astype(np.float64)
    gaussian_kernel = np.zeros((size,size))
    output_img = np.zeros(img.shape)
    for i in range(size):
        for j in range(size):
                gaussian_kernel[i][j]=np.exp(-(((i-n)**2)+((j-n)**2))/(2*(sigmad**2)))
    for i in range(n,img.shape[0]-n):
        for j in range(n,img.shape[1]-n):
            for k in range(3):
                sub = img[i-n:i+n+1,j-n:j+n+1,k:k+1].reshape(size,-1)
                sc = sub-img[i][j][k]
                rc = np.exp(-((sc*sc)/(2*sigmar*sigmar)))
                kernel = gaussian_kernel*rc
                output_img[i][j][k] = (sum(sum(sub*kernel)))/sum(sum(kernel))
    return output_img.astype('uint8')
```

```python
size_sigmad_sigmar = [(7,1.2,2),(9,1.2,2),(11,1.2,2),(7,1.5,40),(9,1.5,30),(11,1.5,20),(11,1.83,80)]
```

```python
img = cv2.imread('../images/mountain.png',cv2.IMREAD_UNCHANGED)
img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(100,50))
plt.subplot(len(size_sigmad_sigmar)+1,1,1)
i = 2
plt.imshow(img,cmap='gray')
plt.title('original image')
for v in size_sigmad_sigmar:
    sigmad = v[1]
    sigmar = v[2]
    size = v[0]
    output_img1 = bilateral_filter(img,sigmad,sigmar,size)
    plt.subplot(len(size_sigmad_sigmar)+1,1,i)
    i = i+1
    plt.imshow(output_img1,cmap='gray')
    plt.title('sigmad = '+str(sigmad)+' sigmar= '+str(sigmar)+' size = '+str(size))
```

# Observations

On increasing the size of the filter the blurring increases.Bilateral filter preserves edges and gaussian does not when smoothing.
