# DFT

```python
def DFT_1D(f):
    f1 = f.copy()
    f2 = np.asarray(f,dtype=complex)
    sum =0 
    for i in range(f1.shape[0]):
        for j in range(f1.shape[0]):
            sum = sum + f2[j]*np.exp(-2j*np.pi*j*i*np.divide(1,f1.shape[0],dtype=complex))
        f1[i] = sum
        sum=0
    return f1
```

```python
def DFT_2D(image):
  img = image.copy()
  m = np.arange(img.shape[0])
  n = np.arange(img.shape[1])
  output_img = np.zeros((img.shape[0],img.shape[1]),complex)
  temp = np.zeros((img.shape[0],img.shape[1]),complex)
  x = m.reshape((img.shape[0],1))
  y = n.reshape((img.shape[1],1))
  for r in range(img.shape[0]):
      temp[r] = np.dot(1j*np.sin(-2*np.pi*y*n/img.shape[1]) + np.cos(-2*np.pi*y*n/img.shape[1]),img[r])
  for c in range(img.shape[1]):
      output_img[:,c] = np.dot(1j*np.sin(-2*np.pi*x*m/img.shape[0]) + np.cos(-2*np.pi*x*m/img.shape[0]),temp[:,c])
  return output_img
```

```python
img = cv2.imread("../images/lena.jpg",cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img,(256,256))
img= np.pad(img,((0,img.shape[0]),(0,img.shape[1])),mode='constant')
img= img.astype('int16')
for r in range(img.shape[0]):
    for c in range(img.shape[1]):
      if (r+c)%2:
        img[r][c] = -1 * img[r][c]
dft_img = DFT_2D(img)
opt1 = np.log(1 + np.abs(dft_img))
plt.imshow(opt1,cmap='gray')
```

# FFT

```python
def FFT_1D(f):
    f1 = f.copy()
    f1 = np.asarray(f1,dtype=complex)
    if f1.shape[0]%2!= 0:
        return 
    if f1.shape[0]<=4:
        return DFT_1D(f1)
    else:
        f1_even = FFT_1D(f1[::2])  # even part
        f1_odd = FFT_1D(f1[1::2])  # odd part
        n = f1.shape[0]
        W = np.exp(-2j*np.pi*np.arange(n)/n)
        k = n//2
        opf = np.concatenate([f1_even + f1_odd * W[:k],f1_even + f1_odd * W[k:]])

    return opf
```

```python
def FFT_2D(f):
    f1 = f.copy()
    opf = np.zeros(f1.shape, dtype=complex)
    if len(f1.shape)==3: #colored image
        for k in range(3):
            opf[:,:,k]=FFT_2D(f1[:,:,k])
    if len(f1.shape) == 2: #grayscale or one channel
        for i in range(f1.shape[0]):opf[i,:]=FFT_1D(f1[i,:])
        for j in range(f1.shape[1]):opf[:,j]=FFT_1D(opf[:,j])
            
    return opf
```

```python
img = cv2.imread("../images/lena.jpg",cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img,(256,256))
img= np.pad(img,((0,img.shape[0]),(0,img.shape[1])),mode='constant')
img= img.astype('int16')
for r in range(img.shape[0]):
    for c in range(img.shape[1]):
      if (r+c)%2:
        img[r][c] = -1 * img[r][c]
fft_img = FFT_2D(img)
opt2 = np.log(1 + np.abs(fft_img))
plt.imshow(opt2,cmap='gray')
```

# IFFT

```python
def IFFT_1D(f):
    f1=f.copy()
    f1 = np.asarray(f1, dtype=complex)
    opf = np.conjugate(FFT_1D(np.conjugate(f1)))/f1.shape[0]
    return opf
```

```python
def IFFT_2D(f):
    f1 = f.copy()
    opf = np.zeros(f1.shape, dtype=complex)
    if len(f1.shape)==3: #colored image
        for k in range(3):
            opf[:,:,k]=IFFT_2D(f1[:,:,k])
    if len(f1.shape) == 2: #grayscale or one channel
        for i in range(f1.shape[0]):opf[i,:]=IFFT_1D(f1[i,:])
        for j in range(f1.shape[1]):opf[:,j]=IFFT_1D(opf[:,j])
    opf=np.real(opf)        
    return opf
```

```python
inverse_img = IFFT_2D(fft_img)
inverse_img = np.real(inverse_img)
img=inverse_img.copy()
for r in range(img.shape[0]):
    for c in range(img.shape[1]):
      if (r+c)%2:
        img[r][c] = -1 * img[r][c]
img=img.astype(np.uint8)
img = img[:img.shape[0]//2,:img.shape[1]//2]
plt.imshow(img,'gray')
```

```python
import time
sizes = [(64,64),(128,128),(256,256),(512,512)]
time1 = []
time2 = []
img = cv2.imread("../images/lena.jpg", cv2.IMREAD_GRAYSCALE)
for x in sizes:
  image = img
  resized_img = cv2.resize(image,x)
  s = time.time()
  optn = DFT_2D(resized_img)
  e = time.time()
  time1.append(e-s)
  s = time.time()
  optn = FFT_2D(resized_img)
  e = time.time()
  time2.append(e-s)
fig = plt.figure(figsize=(10,5))
sizer = [64,128,256,512]
plt.plot(sizer,time1,'g', label="dft")
plt.plot(sizer,time2,'r', label="fft")
plt.legend()
plt.xlabel("Size")
plt.ylabel("time")
plt.title("dft vs fft")
plt.show()
```
