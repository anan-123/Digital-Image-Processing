```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def erosion(image, size):
    img = image.copy()
    kernel = np.ones((size,size), dtype=np.uint8)
    img_op = np.zeros(img.shape)
    n = size // 2
      
    for r in range(n,img.shape[0]-n):
      for c in range(n,img.shape[1]-n):
        for k in range(3):
            img_op[r][c][k] = np.min(img[r-n:r+n+1,c-n:c+n+1,k:k+1]*kernel)
  
    return img_op.astype(np.uint8)

```

```python
def dilation(image, size):
    img = image.copy()
    kernel = np.ones((size,size), dtype=np.uint8)
    img_op = np.zeros(img.shape)
    n = size // 2
      
    for r in range(n,img.shape[0]-n):
      for c in range(n,img.shape[1]-n):
        for k in range(3):
            img_op[r][c][k] = np.max(img[r-n:r+n+1,c-n:c+n+1,k:k+1]*kernel)
  
    return img_op.astype(np.uint8)
```

```python
img = cv2.imread('../images/eye.png',cv2.IMREAD_UNCHANGED)
img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,20))
plt.subplot(1,4,1)
plt.imshow(img)
plt.title('original image')
img1 = erosion(img,13)
img2 = dilation(img,13)
img3 = img2-img1
plt.subplot(1,4,2)
plt.imshow(img1)
plt.title('eroded image')
plt.subplot(1,4,3)
plt.imshow(img2)
plt.title('dilated image')
plt.subplot(1,4,4)
plt.imshow(img3)
plt.title('morphological gradient image')
```

# Explanation


The size of structuring element affects the sharpness of edges: if the structuring element is too large then there is too much erosion and dilation hence very blurred and inaccurate edges in the morphological gradient. If the struturing element is too small then there is no difference in internal and external edge boundaries and there is no difference between other edge detectors. From above we can see that size around 8-15 is a good choice.
